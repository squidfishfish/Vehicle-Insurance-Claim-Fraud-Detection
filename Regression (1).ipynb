```python
#based on the visuliztion analysis, we got these varibales are related to the fraud.
mydata1 = mydata[['AccidentArea', 'Sex','AgeOfPolicyHolder','Fault','VehicleCategory','VehiclePrice','FraudFound_P','PastNumberOfClaims','BasePolicy']]
mydata1
```

```python
# creat new cloumns name
col_name = mydata1.columns.tolist()  
col_name.insert(9,'Rural')
col_name.insert(10,'Male')
col_name.insert(11,'Policy Holder Fault')
col_name.insert(12,'Sedan')
col_name.insert(13,'Frist Claim')
col_name.insert(14,'Collision Coverage')
print(col_name)
```

```python
# Creat dummy variables cloumns
DF1=mydata1.reindex(columns=col_name)
DF1.loc[DF1.AccidentArea == "Rural", 'Rural'] = 1
DF1.loc[DF1.Sex == "Male", 'Male'] = 1
DF1.loc[DF1.Fault == "Policy Holder", 'Policy Holder Fault'] = 1
DF1.loc[DF1.VehicleCategory == "Sedan", 'Sedan'] = 1
DF1.loc[DF1.PastNumberOfClaims == "none", 'Frist Claim'] = 1
DF1.loc[DF1.BasePolicy != "Liability", 'Collision Coverage'] = 1
DF1.loc[DF1.VehiclePrice == "more than 69000", 'VP_69000'] = 1
DF1.loc[DF1.VehiclePrice == "less than 20000", 'VP_20000'] = 1
DF1.loc[DF1.BasePolicy != "Liability", 'Collision Coverage'] = 1
DF1 = DF1.fillna(0)
DF1
```

```python
# Based on the age analysis, we have three groups of age.
DF1.loc[(DF1.AgeOfPolicyHolder == '16 to 17') , 'Under 35'] = 1
DF1.loc[(DF1.AgeOfPolicyHolder == '18 to 20') , 'Under 35'] = 1
DF1.loc[(DF1.AgeOfPolicyHolder == '21 to 25') , 'Under 35'] = 1
DF1.loc[(DF1.AgeOfPolicyHolder == '26 to 30') , 'Under 35'] = 1
DF1.loc[(DF1.AgeOfPolicyHolder == '31 to 35') , 'Under 35'] = 1
DF1.loc[(DF1.AgeOfPolicyHolder == '36 to 40') , 'Under 35'] = 0
DF1.loc[(DF1.AgeOfPolicyHolder == '41 to 50') , 'Under 35'] = 0
DF1.loc[(DF1.AgeOfPolicyHolder == '51 to 65') , 'Under 35'] = 0
DF1.loc[(DF1.AgeOfPolicyHolder == 'over 65') , 'Under 35'] = 0
DF1
```
